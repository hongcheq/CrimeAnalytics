### Read data

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

crime_df = pd.read_csv("./assignment6/sanfrancisco_incidents_summer_2014.csv")

crime_df
```

### Step 1: count the number of crime incidents by category

```python
crime_cate = crime_df['Category']
crime_cate_ct = crime_cate.value_counts().rename_axis('Descript').reset_index(name='counts')
crime_cate_ct
```

```python
# plot
fig, axes = plt.subplots(1,1,figsize=(12,10))
sns.barplot(x='counts', y='Descript', data=crime_cate_ct, ax=axes)
plt.xlabel('')
plt.title('Number of crime incidents of different category')
plt.tight_layout()
plt.show()
```

#### During 2014 summer, the most frequent crime incidents in San Francisco is larceny/theft. Other types of crime incidents, such as non-criminal, assalt, and vehicle theft, are also quite common. The pornography/obscene mat, bribery, and gambling categories of crime incidents have the least reported number.

### Step 2: count the number of most frequent crime incident ("larceny/theft") by different district

```python
crime_cate = crime_df['Category']
crime_cate_ct = crime_cate.value_counts().rename_axis('Descript').reset_index(name='counts')
crime_cate_ct
```

####

```python
crime_dist = crime_df.loc[crime_df.Category=='LARCENY/THEFT']['PdDistrict']
crime_dist_ct = crime_dist.value_counts().rename_axis('PdDistrict').reset_index(name='counts')
crime_dist_ct
```

```python
# plot
fig, axes = plt.subplots(1,1,figsize=(12,10))
sns.barplot(x='PdDistrict', y='counts', data=crime_dist_ct, ax=axes)
plt.xlabel('')
plt.title('Number of larceny/theft incidents in different districts')
plt.tight_layout()
plt.show()
```

#### During 2014 summer, most of the larceny/theft incidents occur in the Southern District with over 2500 cases. This is followed by the Central District and the Northern District.

### Step 3: inspect the most frequent crime incident ("larceny/theft") by different day of the week.

```python
crime_df['DayOfWeek'] = pd.Categorical(crime_df['DayOfWeek'], categories=\
                                   ['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday', 'Sunday'],\
                                   ordered=True)

crime_dow = crime_df.loc[crime_df.Category=='LARCENY/THEFT']['DayOfWeek']
crime_dow = crime_dow.value_counts()
crime_dow_sort = crime_dow.sort_index().rename_axis('DayOfWeek').reset_index(name='counts')
crime_dow_sort
```

```python
# plot
fig, axes = plt.subplots(1,1,figsize=(12,10))
sns.barplot(x='DayOfWeek', y='counts', data=crime_dow_sort, ax=axes)
plt.xlabel('DayOfWeek')
plt.title('Number of larceny/theft incidents on different day of the week')
plt.tight_layout()
plt.show()
```

#### During 2014 summer, the number of cases in larceny/theft incidents occur most often on Saturday, followed by Sunday and Friday.

### Step 4: inspect the most frequent crime incident ("larceny/theft") by different hour of the day.

```python
crime_diurnal = crime_df.loc[crime_df.Category=='LARCENY/THEFT']['Time']
crime_diurnal_hr = pd.to_datetime(crime_diurnal).dt.hour
crime_diurnal_hr_ct = crime_diurnal_hr.value_counts()
crime_diurnal_hr_ct_sort_tmp = crime_diurnal_hr_ct.sort_index()
crime_diurnal_hr_ct_sort = crime_diurnal_hr_ct_sort_tmp.rename_axis('hour').reset_index(name='counts')
crime_diurnal_hr_ct_sort
```

```python
# plot
fig, axes = plt.subplots(1,1,figsize=(12,10))
sns.barplot(x='hour', y='counts', data=crime_diurnal_hr_ct_sort, ax=axes)
plt.xlabel('Hour of day')
plt.title('Number of larceny/theft incidents on different hour of the day')
plt.tight_layout()
plt.show()
```

#### During 2014 summer, the larceny/theft incidents occurr most often at evening hours (17:00-22:00). The occurrance rate is also pretty high at noon (12:00).
